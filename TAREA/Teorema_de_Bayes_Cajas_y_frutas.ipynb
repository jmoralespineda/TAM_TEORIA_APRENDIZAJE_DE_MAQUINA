# El ejercicio consiste en 2 cajas de un color cada una una red y otra blue. En la caja red hay 8 frutas siendo 6 naranjas y 2 manzanas. La caja blue tiene 4 frutas siendo 3 manzanas y 1 naranja. Hay 2 variables aleatorias siendo x la variable aleatoria caja y Y la variable aleatoria fruta. Se determinó que la probabilidad de la caja roja es 0.4 y la probabilidad de la caja azul es de 0.6. Puedes ayudarme a realizar una simulación en donde se evidencie este problema de probabilidades? Tambien agrega el calculo de algunas probabilidades condicionales.

```python
import random
from collections import Counter

# -----------------------------
# Definir el contenido de cada caja
# -----------------------------
red_box = ['naranja'] * 6 + ['manzana'] * 2   # 8 frutas
blue_box = ['naranja'] * 1 + ['manzana'] * 3  # 4 frutas

# -----------------------------
# Lista de cajas y sus probabilidades
# -----------------------------
cajas = ['red', 'blue']
prob_cajas = [0.4, 0.6]  # Suma = 1.0

# -----------------------------
# Diccionario de contenido de cajas
# -----------------------------
contenido_cajas = {
    'red': red_box,
    'blue': blue_box
}

# -----------------------------
# Simulación
# -----------------------------
num_simulaciones = 10000
resultados = []

for _ in range(num_simulaciones):
    caja = random.choices(cajas, weights=prob_cajas, k=1)[0]
    fruta = random.choice(contenido_cajas[caja])
    resultados.append((caja, fruta))

# -----------------------------
# Conteos
# -----------------------------
conteo_caja_fruta = Counter(resultados)
conteo_frutas = Counter([r[1] for r in resultados])
conteo_cajas = Counter([r[0] for r in resultados])

# -----------------------------
# Resultados generales
# -----------------------------
print("=== Distribución conjunta (caja, fruta) ===")
for clave, valor in conteo_caja_fruta.items():
    prob = valor / num_simulaciones
    print(f"{clave}: {valor} veces → Probabilidad: {prob:.4f}")

print("\n=== Distribución marginal por fruta ===")
for fruta, valor in conteo_frutas.items():
    print(f"{fruta}: {valor} veces → {valor / num_simulaciones:.4f}")

print("\n=== Distribución marginal por caja ===")
for caja, valor in conteo_cajas.items():
    print(f"{caja}: {valor} veces → {valor / num_simulaciones:.4f}")

# -----------------------------
# Probabilidades condicionales
# -----------------------------

# 1. P(fruta = manzana | caja = red)
manzana_y_red = conteo_caja_fruta[('red', 'manzana')]
total_red = conteo_cajas['red']
p_manzana_dado_red = manzana_y_red / total_red
print(f"\nP(manzana | red) = {p_manzana_dado_red:.4f}")

# 2. P(fruta = naranja | caja = blue)
naranja_y_blue = conteo_caja_fruta[('blue', 'naranja')]
total_blue = conteo_cajas['blue']
p_naranja_dado_blue = naranja_y_blue / total_blue
print(f"P(naranja | blue) = {p_naranja_dado_blue:.4f}")

# 3. P(caja = blue | fruta = manzana)
manzanas_total = conteo_frutas['manzana']
manzana_y_blue = conteo_caja_fruta[('blue', 'manzana')]
p_blue_dado_manzana = manzana_y_blue / manzanas_total
print(f"P(blue | manzana) = {p_blue_dado_manzana:.4f}")




```

# Problema modificado: El cambio consiste en que en vez de ser 3 cajas, ahora hay 3, en donde la caja 3 se llamará Green. Además, también se agrego otra fruta la cual será el banano. La distribucion de las frutas es el siguente: En la caja red hay 4 naranjas, 2 manzanas y 1 banano. En la caja blue hay 1 naranja, 3 manzanas y 1 banano. En la caja green hay 2 naranjas, 2 manzanas y 4 bananos. La probabilidad de la caja red es de 0.35, la de la caja blue es de 0.45 y el de a caja green es de 0.20. Tambien agrega el cálculo de algunas probabilidades condicionales.

```python
import random
from collections import Counter

# -----------------------------
# Definir el contenido de cada caja
# -----------------------------
red_box = ['naranja'] * 4 + ['manzana'] * 2 + ['banano'] * 1     # 7 frutas
blue_box = ['naranja'] * 1 + ['manzana'] * 3 + ['banano'] * 1    # 5 frutas
green_box = ['naranja'] * 2 + ['manzana'] * 2 + ['banano'] * 4   # 8 frutas

# -----------------------------
# Lista de cajas y sus probabilidades
# -----------------------------
cajas = ['red', 'blue', 'green']
prob_cajas = [0.35, 0.45, 0.20]  # Suma = 1.0

# -----------------------------
# Diccionario para acceder a las frutas según la caja elegida
# -----------------------------
contenido_cajas = {
    'red': red_box,
    'blue': blue_box,
    'green': green_box
}

# -----------------------------
# Inicializar lista para guardar resultados de simulación
# -----------------------------
resultados = []

# -----------------------------
# Cantidad de veces que vamos a simular el experimento
# -----------------------------
num_simulaciones = 10000

# -----------------------------
# Simulación del experimento
# -----------------------------
for _ in range(num_simulaciones):
    # Elegir una caja según las probabilidades definidas
    caja = random.choices(cajas, weights=prob_cajas, k=1)[0]

    # Elegir una fruta de la caja seleccionada (al azar)
    fruta = random.choice(contenido_cajas[caja])

    # Guardar el resultado como una tupla (caja, fruta)
    resultados.append((caja, fruta))

# -----------------------------
# Análisis de los resultados
# -----------------------------

# Conteo combinado de (caja, fruta)
conteo_caja_fruta = Counter(resultados)

# Conteo total por fruta
conteo_frutas = Counter([r[1] for r in resultados])

# Conteo total por caja
conteo_cajas = Counter([r[0] for r in resultados])

# -----------------------------
# Mostrar resultados
# -----------------------------
print("=== Conteo por combinación (caja, fruta) ===")
for clave, valor in conteo_caja_fruta.items():
    prob = valor / num_simulaciones
    print(f"{clave}: {valor} veces → Probabilidad aproximada: {prob:.4f}")

print("\n=== Distribución total por fruta ===")
for fruta, valor in conteo_frutas.items():
    prob = valor / num_simulaciones
    print(f"{fruta}: {valor} veces → Probabilidad: {prob:.4f}")

print("\n=== Distribución total por caja ===")
for caja, valor in conteo_cajas.items():
    prob = valor / num_simulaciones
    print(f"{caja}: {valor} veces → Probabilidad: {prob:.4f}")

# Probabilidades condicionales
# 1. P(fruta = manzana | caja = blue)
manzana_y_blue = conteo_caja_fruta[('blue', 'manzana')]
total_blue = conteo_cajas['blue']
p_manzana_dado_blue = manzana_y_blue / total_blue
print(f"\nP(manzana | blue) = {p_manzana_dado_blue:.4f}")

# 2. P(caja = red | fruta = naranja)
# Total naranjas en todas las cajas
total_naranjas = conteo_frutas['naranja']

# De esas, ¿cuántas vinieron de red?
naranja_y_red = conteo_caja_fruta[('red', 'naranja')]
p_red_dado_naranja = naranja_y_red / total_naranjas
print(f"P(red | naranja) = {p_red_dado_naranja:.4f}")

# 3. P(banano | green)
banano_y_green = conteo_caja_fruta[('green', 'banano')]
total_green = conteo_cajas['green']
p_banano_dado_green = banano_y_green / total_green
print(f"P(banano | green) = {p_banano_dado_green:.4f}")

```
