```python
import numpy as np
import matplotlib.pyplot as plt
M = 1000 # numero de simulaciones de montecarlo
N = 100
A = 5
A1 = np.zeros((M,1))
A2 = np.zeros((M,1))
for m in range(M): #1 iteracion de montecarlo
    print(f'simulacion{m+1}/{M} de montecarlo...')
    w = np.random.randn(N) # me estregan un nuevo set de datos
    x = A + w
    A1[m] = np.mean(x) #estimador 1 -> promedio iid verosimilitud por Gaussiana
    A2[m] = x[0] # estimador 2 -> promedio = muestra 1

plt.hist(A1,bins=20,label='estimador 1',alpha=0.5)
plt.hist(A2,bins=20,label='estimador 2',alpha=0.5)
plt.legend()
plt.show()

#w= np.random.randn(N)
#x = A + w
plt.plot(w,label='w')
plt.plot(x,label='x')
plt.legend()
plt.show()
```

```python
mw = np.mean(w)
mx = np.mean(x)
vw = np.var(w)
vx = np.var(x)
print(mw,mx)
print(vw,vx)
```
